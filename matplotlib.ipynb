```python
from matplotlib import pyplot as plt
import numpy as np
import matplotlib
matplotlib.rcParams['font.sans-serif'] = ['SimHei']  #识别文字的代码
```

```python
x = np.arange(10)
y = np.random.randn(10)
plt.xlabel('x轴')
plt.ylabel('y轴')
plt.rcParams['axes.unicode_minus']=False   #用来显示中文
plt.plot(x,y) #折线图
plt.bar(x,y)  #柱状图
plt.scatter(x,y)  #散点图
plt.show()
```

```python
labels = ['期中-笔试','期中-机试','期末-笔试','期末-机试']
data=[10,20,10,30]
explode=[0,0,0,0.1]  #explode板块的间隔；startangle旋转的角度数和正反；autopct保留的百分比；
plt.pie(data,labels=labels,autopct='%1.1f%%',shadow=False,startangle=-90,explode=explode)
plt.show()
```
